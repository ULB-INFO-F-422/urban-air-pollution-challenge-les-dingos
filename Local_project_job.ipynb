### Step 3 : Missing Value Imputation
#### 3.1 Assessing Missingness

```r
# Calculate the percentage of missing values for each column
missing_percent <- sapply(train_data, function(x) sum(is.na(x)) / length(x) * 100)

# Sort and print the missing values percentage
missing_percent_sorted <- sort(missing_percent, decreasing = TRUE)
print(missing_percent_sorted)

# Plot the missing data percentages for a visual overview
library(ggplot2)
missing_data_frame <- data.frame(Feature = names(missing_percent_sorted), MissingPercentage = missing_percent_sorted)
ggplot(missing_data_frame, aes(x = reorder(Feature, -MissingPercentage), y = MissingPercentage)) +
  geom_bar(stat = "identity") +
  theme_minimal() +
  labs(x = "Features", y = "Percentage Missing Data", title = "Missing Data by Feature") +
  coord_flip()  # Flips the axes for better readability
```

#### 3.2 Dropping Columns that have more than 30% of missing data

Personnal choice after looking at the "Missing Data by Feature" chart 

```r
# Identify columns to retain (those with less than 30% missing data)
cols_to_keep <- names(train_data)[missing_percent < 30]

# Select these columns from the data
data_cleaned <- train_data[, cols_to_keep]

# Optionally, print out the structure or summary of the cleaned data to confirm
print(summary(data_cleaned))
str(data_cleaned)

# If you want to visualize the remaining missing data:
library(ggplot2)
missing_data_frame_cleaned <- data.frame(Feature = names(data_cleaned),
                                         MissingPercentage = sapply(data_cleaned, function(x) sum(is.na(x)) / length(x) * 100))
ggplot(missing_data_frame_cleaned, aes(x = reorder(Feature, -MissingPercentage), y = MissingPercentage)) +
  geom_bar(stat = "identity", fill = "steelblue") +
  theme_minimal() +
  labs(x = "Features", y = "Percentage Missing Data", title = "Remaining Missing Data by Feature After Cleaning") +
  coord_flip()
```

```r
#writting to a csv file to view the data frame in excel sheet for a clearer view
write.csv(data_cleaned, "30%Cleaned_data.csv", row.names=FALSE)
```

```r
# Transforming the values 0 into NA since by visual contemplation, it seems they are the same
data_no0 <- data_cleaned
data_no0[data_cleaned == 0] <- NA
```

```r
missing_data_frame_cleaned_no0 <- data.frame(Feature = names(data_no0),
                                         MissingPercentage = sapply(data_no0, function(x) sum(is.na(x)) / length(x) * 100))
ggplot(missing_data_frame_cleaned, aes(x = reorder(Feature, -MissingPercentage), y = MissingPercentage)) +
  geom_bar(stat = "identity", fill = "steelblue") +
  theme_minimal() +
  labs(x = "Features", y = "Percentage Missing Data", title = "Remaining Missing Data by Feature After Cleaning") +
  coord_flip()
```

```r
# Calculate the total number of missing values in the data frame
total_missing_values <- sum(is.na(data_no0))

# Calculate the total number of data points in the data frame
total_data_points <- prod(dim(data_no0))

# Calculate the percentage of missing data in the entire data frame
overall_percentage_missing <- (total_missing_values / total_data_points) * 100

# Print the result
print(paste("Overall percentage of missing data:", overall_percentage_missing, "%"))
```
